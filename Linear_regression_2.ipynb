```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

similar like simple linear regression

now we are reading multiple fields of x,

taking $x_{0}$ = 1 for every sample and building matrix $X$

```python
X = []
Y = []

#prepare matrix X and Y
for line in open('machine_learning_examples-master/linear_regression_class/data_2d.csv'):
    x1, x2, y = line.split(',')
    X.append([1, float(x1), float(x2)])
    Y.append(float(y))

#convert into numpy array
X = np.array(X)
Y = np.array(Y)

#scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X[:,1], X[:,2], Y)
ax.set_xlabel('X0')
ax.set_ylabel('X1')
ax.set_zlabel('y')
plt.show
```

Calculate weights

to find $\hat{y}$ we need to find $\theta$

\begin{align}
\theta  = (X^TX)^{-1} X^T\vec{y}
\end{align}

or solve 
\begin{align}
(X^TX)\theta = X^T\vec{y}
\end{align}

once we get $\theta$ we solve for $\hat{y}$

given $$ \hat{y} = \theta_{0} x_{0} + \theta_{1} x_{1} + \theta_{1} x_{1} $$

$$x_{0} = 1$$

same as $$  \hat{y} = \sum_{i=0}^{2}\theta _{i}x_{i} = (X . \theta)$$


```python
#solve equ Ax=b
theta = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))

#find yhat = (X.0)
Yhat = np.dot(X, theta)
```

compute r-squared

```python
d1 = Y - Yhat #sum total =sstot
d2 = Y - Y.mean() #sum mean = ssmena
r2 = 1- (d1.dot(d1)/d2.dot(d2)) # #r squared estimation = 1 - square of (sum_total / sum_mean)
print(r2)
```
